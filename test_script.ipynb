```python
import random
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import time

min_number = 1
max_number = 37

def addition_questions():
    number_1 = random.randint(min_number, max_number)
    number_2 = random.randint(min_number, max_number)
    question = f'{number_1} + {number_2}'
    correct_answer = number_1 + number_2
    return question, correct_answer

def subtraction_questions():
    number_1 = random.randint(min_number, max_number)
    number_2 = random.randint(min_number, max_number)
    question = f'{number_1} - {number_2}'
    correct_answer = number_1 - number_2
    return question, correct_answer

def mixed_questions_1():
    number_1 = random.randint(min_number, max_number)
    number_2 = random.randint(min_number, max_number)
    number_3 = random.randint(min_number, max_number)
    question = f'{number_1} - {number_2} + {number_3}'
    correct_answer = number_1 - number_2 + number_3
    return question, correct_answer

def mixed_questions_2():
    number_1 = random.randint(min_number, max_number)
    number_2 = random.randint(min_number, max_number)
    number_3 = random.randint(min_number, max_number)
    question = f'{number_1} + {number_2} - {number_3}'
    correct_answer = number_1 + number_2 - number_3
    return question, correct_answer

def main():
    maths_test_start = widgets.HTML(value='For this part of the test, you will be asked to answer a number of Maths questions. Your score will be displayed at the end of the test. The test will start in a few seconds.')
    display(maths_test_start)
    time.sleep(5)
    clear_output(wait=False)

    score_1, score_2, score_3, score_4 = 0, 0, 0, 0
    total_time_taken_1, total_time_taken_2, total_time_taken_3, total_time_taken_4 = 0, 0, 0, 0

    for i in range(10):  
        question, correct_answer = addition_questions()
        start_time=time.time()
        user_answer = input(question)
        end_time=time.time()
        time_taken_1=end_time-start_time
        total_time_taken_1 += time_taken_1
        try:
            user_numerical_answer = int(user_answer)
            if user_numerical_answer == correct_answer:
                print("Correct!\n")
                print(f'Time taken was {time_taken_1:.2f} seconds')
                score_1 += 1
                clear_output(wait=True)
            else:
                print(f"Wrong. The correct answer is {correct_answer}\n")
                clear_output(wait=True)
        except ValueError:
            print("Invalid input. Please enter a valid number.\n")
            clear_output(wait=True)

    for i in range(10):  
        question, correct_answer = subtraction_questions()
        start_time=time.time()
        user_answer = input(question)
        end_time=time.time()
        time_taken_2=end_time-start_time
        total_time_taken_2 += time_taken_2
        try:
            user_numerical_answer = int(user_answer)
            if user_numerical_answer == correct_answer:
                print("Correct!\n")
                print(f'Time taken was {time_taken_2:.2f} seconds')
                score_2 += 1
                clear_output(wait=True)
            else:
                print(f"Wrong. The correct answer is {correct_answer}\n")
                clear_output(wait=True)
        except ValueError:
            print("Invalid input. Please enter a valid number.\n")
            clear_output(wait=True)

    for i in range(5):  
        question, correct_answer = mixed_questions_1()
        start_time=time.time()
        user_answer = input(question)
        end_time=time.time()
        time_taken_3=end_time-start_time
        total_time_taken_3 += time_taken_3
        try:
            user_numerical_answer = int(user_answer)
            if user_numerical_answer == correct_answer:
                print("Correct!\n")
                print(f'Time taken was {time_taken_3:.2f} seconds')
                score_3 += 1
                clear_output(wait=True)
            else:
                print(f"Wrong. The correct answer is {correct_answer}\n")
                clear_output(wait=True)
        except ValueError:
            print("Invalid input. Please enter a valid number.\n")
            clear_output(wait=True)

    for i in range(5): 
        question, correct_answer = mixed_questions_2()
        start_time=time.time()
        user_answer = input(question)
        end_time=time.time()
        time_taken_4=end_time-start_time
        total_time_taken_4 += time_taken_4
        try:
            user_numerical_answer = int(user_answer)
            if user_numerical_answer == correct_answer:
                print("Correct!\n")
                print(f'Time taken was {time_taken_4:.2f} seconds')
                score_4 += 1
                clear_output(wait=True)
            else:
                print(f"Wrong. The correct answer is {correct_answer}\n")
                clear_output(wait=True)
        except ValueError:
            print("Invalid input. Please enter a valid number.\n")
            clear_output(wait=True)
    global final_score
    global total_time
    final_score = score_1 + score_2 + score_3 + score_4
    total_time = total_time_taken_1 + total_time_taken_2 + total_time_taken_3 + total_time_taken_4
    print(total_time)
    print(f"You got {score_1 + score_2 + score_3 + score_4} out of 30 questions correct!")


main()
```

```python
from IPython.display import Image, display, clear_output
import ipywidgets as widgets
from threading import Timer
import time

example_exercise_text = widgets.HTML(value='''
    You will now be asked to answer an example question to familiarize yourself with the task.
    The image will be shown for a brief period of time, after which you will have a few seconds to pick the side which you believe contained MORE circles.
    To start, click 'Begin'
    '''
)

display(example_exercise_text)

output_img103 = widgets.Output()
output_img101 = widgets.Output()

img103 = Image(filename='103.png')
img101 = Image(filename='101.png')

output = widgets.Output()

begin_button = widgets.Button(description='Begin')
left_button = widgets.Button(description='Left')
right_button = widgets.Button(description='Right')
left_right_box = widgets.HBox([left_button, right_button])
left_right_box.layout.display = 'none'

time_1 = None

def begin_clicked(b):
    global time_1
    with output_img103:
        example_exercise_text.layout.display = 'none'
        begin_button.layout.display = 'none'
        display(img103) 
        time.sleep(0.75)
        clear_output(wait=False)
    with output_img101:
        time.sleep(0.75)
        display(img101)
        left_right_box.layout.display = ''
        time_1 = time.time()

def left_right_clicked_fx(b):
    with output_img101:
        time_2 = time.time()
        time_taken = time_2 - time_1
        left_right_box.layout.display = 'none'
        clear_output(wait=True)
        print(f'Selected button: {b.description}')
        print('Time taken was', time_taken)

begin_button.on_click(begin_clicked)
left_button.on_click(left_right_clicked_fx)
right_button.on_click(left_right_clicked_fx)
            
display(widgets.VBox([begin_button, output_img103, output_img101, left_right_box]))
```

```python
import os
from IPython.display import Image, display, HTML, clear_output
import ipywidgets as widgets
import time

def run_experiment(img_list, img101_path):
    global iteration
    iteration = iter(img_list)
    img101_displayed = False 
    display_next(img101_path, img101_displayed)

def display_next(img101_path, img101_displayed):
    try:
        img_path = next(iteration)
        img_name = os.path.basename(img_path) 

        with output:
            clear_output(wait=True)
            display(Image(img_path))

        time.sleep(1.5)

        start_time = time.time()

        left_button = widgets.Button(description='Left')
        right_button = widgets.Button(description='Right')

        user_choices = {'image_path': img_path, 'choice': None}

        def left_clicked(b):
            user_choices['choice'] = 'Left'
            end_time = time.time()
            response_time = end_time - start_time
            update_html(img_name, user_choices, response_time)
            time.sleep(2)
            display_next(img101_path, True)

        def right_clicked(b):
            user_choices['choice'] = 'Right'
            end_time = time.time()
            response_time = end_time - start_time
            update_html(img_name, user_choices, response_time)
            time.sleep(2)
            display_next(img101_path, True)

        left_button.on_click(left_clicked)
        right_button.on_click(right_clicked)

        with output:
            clear_output(wait=True)
            display(Image(img101_path)) 
            display(widgets.HBox([left_button, right_button]))

    except StopIteration:
        print("Experiment completed.")

output_html_list = []

def update_html(img_name, user_choices, response_time):
    global output_html
    global output_html_list
    name_html = f'<p>Image Name: {img_name}</p>'
    choice_html = f'<p>You chose: {user_choices["choice"]}</p>'
    time_html = f'<p>Response time: {response_time:.2f} seconds</p>'
    output_html = f'{name_html}{choice_html}{time_html}'
    output_html_list.append(output_html)

    with output:
        clear_output(wait=True)
        display(HTML(output_html))

def main():
    img_folder = 'fig_adj'
    img_list = [os.path.join(img_folder, f) for f in os.listdir(img_folder)]
    img101_path = '101.png'

    begin_button = widgets.Button(description='Begin')
    test_start = widgets.HTML(value='To start, press Begin')
    global output
    output = widgets.Output()

    display(begin_button)
    display(test_start)
    display(output)

    def begin_clicked(b):
        begin_button.layout.display = 'none'
        test_start.layout.display = 'none'
        run_experiment(img_list, img101_path)

    begin_button.on_click(begin_clicked)

main()
```
